```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import re
import json
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
```

```python
# Path to driver

path_to_web_driver = "C:/Users/anais.chalant/Desktop/Formation/Certification_DE/chromedriver"
```

```python
# Browser options

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--verbose')
browser = webdriver.Chrome(executable_path=path_to_web_driver,
options=chrome_options)
```

```python
# Website

a = "https://fr.trustpilot.com"
```

```python
# Increment over all pages to get the url for each store

numero = list(range(2,170))
num = [str(i) for i in numero]
num =['?page=' +s for s in num]
num.insert(0,'')

url ="https://fr.trustpilot.com/categories/shopping_fashion" 

ref_stores = []
t1=time.time()
for j in num:
    page = url+j
    Page = browser.get(page)
    try:
        URL = browser.find_elements(By.XPATH, '//*[contains(concat( " ", @class, " " ), concat( " ", "styles_linkWrapper__UWs5j", " " ))]')
        for i in URL: 
            url_company = i.get_attribute('href')
            ref_stores.append(url_company) 
    except:
        pass
t2=time.time() 
print("\nLa complétion du programme prend : ", (t2-t1)/60,"min \n")
```

```python
len(ref_stores)
```

```python
# set up timer
t1=time.time()


# lists
name_store = []
reviews_store = []
score_store = []
type_store = []
five_stars = []
four_stars = []
three_stars = []
two_stars = []
one_star = []


for i in ref_stores:
    print(i)
    Page = browser.get(i)
    
    # Name store
    try:
        name = browser.find_element(By.XPATH, '//*[@id="business-unit-title"]/h1/span[1]')
        name_store.append(name.text.strip())
    except:
        name_store.append('NA')
        
    # Get number of views
    try:
        reviews = browser.find_element(By.XPATH, '//*[@id="business-unit-title"]/span/span')
        reviews_store.append(reviews.text.split('•')[0].strip().replace(u"\u202f", ""))
    except:
        reviews_store.append("NA")
        
    # Get score for the store
    try:
        score = browser.find_element(By.XPATH, '//*[@id="business-unit-title"]/div/div/p')
        score_store.append(score.text)
    except:
        score_store.append("NA")
        
    # Get type of store
    try: 
        size = len(browser.find_elements(By.XPATH, '//*[contains(concat( " ", @class, " " ), concat( " ", "breadcrumb_breadcrumbLink__p1PMo", " " ))]'))
        type_ = browser.find_element(By.XPATH, f'//*[@id="__next"]/div/div/div/main/div/div[1]/nav/ol/li[{size-1}]')
        type_store.append(type_.text)

    except:
        type_store.append("NA")
        
    # Get percentage of 5 stars ratings, 4 stars....
    try:
        rating = browser.find_elements(By.XPATH, '//*[contains(concat( " ", @class, " " ), concat( " ", "styles_percentageCell__cHAnb", " " ))]')        
        five_stars.append(rating[0].text.split(' ')[0])
        four_stars.append(rating[1].text.split(' ')[0])
        three_stars.append(rating[2].text.split(' ')[0])
        two_stars.append(rating[3].text.split(' ')[0])
        one_star.append(rating[4].text.split(' ')[0])

    except: 
        five_stars.append("NA")
        four_stars.append("NA")
        three_stars.append("NA")
        two_stars.append("NA")
        one_star.append("NA")

        
# Create dataframe

df = pd.DataFrame({"Name":name_store, "NumberOfReviews":reviews_store, "Score":score_store,"Category":type_store,
                  "FiveStarsPerc":five_stars,"FourStarsPerc":four_stars,"ThreeStarsPerc":three_stars,"TwoStarsPerc":two_stars,
                  "OneStarPerc":one_star})

t2=time.time() 
print("\nLa complétion du programme prend : ", (t2-t1)/60,"min \n")
```

```python
# Increment over all pages to get the url for each store

numero = list(range(2,1166))
num = [str(i) for i in numero]
num =['?page=' +s for s in num]
num.insert(0,'')

url ="https://fr.trustpilot.com/review/www.fitnessboutique.fr" 


t1=time.time()
finallist = []
for j in num:
    page = url+j
    print(j)
    Page = browser.get(page)
    
    NumReviews =  len(browser.find_elements(By.TAG_NAME,'article'))
    seq = list(range(4,NumReviews+5))
    
    for i in seq:
        try:
            
            name_reviewer = browser.find_element(By.XPATH, f'/html/body/div[1]/div/div/div/main/div/div[4]/section/div[{i}]/article/aside/div/a/span')
            title_review = browser.find_element(By.XPATH, f'/html/body/div[1]/div/div/div/main/div/div[4]/section/div[{i}]/article/section/div[2]/a/h2')
            review_core= browser.find_element(By.XPATH, f'/html/body/div[1]/div/div/div/main/div/div[4]/section/div[{i}]/article/section/div[2]/p[1]')
            date_review= browser.find_element(By.XPATH, f'/html/body/div[1]/div/div/div/main/div/div[4]/section/div[{i}]/article/section/div[2]/p[2]')
            response_review = browser.find_element(By.XPATH, f'/html/body/div[1]/div/div/div/main/div/div[4]/section/div[{i}]/article/div[2]/div[2]/p')
            
            name=name_reviewer.text
            title=title_review.text
            review=review_core.text
            date=date_review.text.split(": ")[1]
            response=response_review.text

        except NoSuchElementException:
            name="NA"
            title="NA"
            review="NA"
            date="NA"
            response="NA"
    
        reviewlist = [{'name':name,'title':title,'review':review, 'date':date, 'response':response}]
        finallist.extend(reviewlist)
        
t2=time.time() 
print("\nLa complétion du programme prend : ", (t2-t1)/60,"min \n")

    
```
